```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.head()
```

```python
# Creating column age_group


labels = ['young_adult', 'adult', 'middle_aged', 'senior']
bins = [0, 25, 40, 60, np.inf]
df['age_group'] = pd.cut(df['age'], labels = labels, bins = bins)
```

```python
df.age_group
```

```python
#Creating column purchase_frequency_in_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Annually' : 365,
    'Every 3 Months' : 90,
    'Weekly' : 7
}

df['purchase_frequency_in_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
